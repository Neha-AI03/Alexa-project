```python
import speech_recognition as sr
import wikipedia

# Initialize the recognizer
listener = sr.Recognizer()

def talk(text):
    print(text)
    # You can integrate pyttsx3 here if you want it to speak out loud:
    # engine.say(text)
    # engine.runAndWait()

def take_command():
    try:
        with sr.Microphone() as source:
            print('Listening...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice)
            command = command.lower()
            if 'alexa' in command:
                command = command.replace('alexa', '')
                print(f"Recognized command: {command}")
                return command
    except Exception as e:
        print("Error:", e)
    return ""

def run_alexa():
    command = take_command()
    if command:
        try:
            info = wikipedia.summary(command, sentences=1)
            talk(info)
        except wikipedia.exceptions.DisambiguationError as e:
            talk("That term is too ambiguous. Try to be more specific.")
        except wikipedia.exceptions.PageError:
            talk("I couldn't find any information on that.")
        except Exception as e:
            talk("Something went wrong.")

# Example usage
run_alexa()
```
